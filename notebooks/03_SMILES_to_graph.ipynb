```python
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors
from IPython.display import display
```

```python
def smiles_to_graph(smiles): 
    molecule = Chem.MolFromSmiles(smiles)
    if not molecule: 
        return None
    formula = rdMolDescriptors.CalcMolFormula(molecule)
    print(f"Processing Molecule: {formula}")

    for atom in molecule.GetAtoms():
        atom.SetProp('atomLabel', f"{atom.GetSymbol()}_{atom.GetIdx()}")

    display(molecule)
    print(f"{'Atom Index':<12} | {'Symbol':<8} | {'Aromatic':<10} | {'Degree':<8}")
    print("-" * 45)
  

    for atom in molecule.GetAtoms():
        index = atom.GetIdx()
        symbol = atom.GetSymbol()
        is_aromatic = atom.GetIsAromatic()
        degree = atom.GetDegree()

        print(f"{index:<12} | {symbol:<8} | {str(is_aromatic):<10} | {degree:<8}")

# Example: Fludarabine
fludarabine_smiles = "C1=NC2=C(N1)C(=NC(=N2)F)N[C@H]3[C@@H]([C@H]([C@H](O3)CO)O)O"
smiles_to_graph(fludarabine_smiles)    

```
